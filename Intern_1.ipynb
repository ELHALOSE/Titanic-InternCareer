# **PreProcessing**


```python
train["Age"]=train["Age"].fillna(method="backfill")
test["Age"]=test["Age"].fillna(method="backfill")
```

```python
print(train.isnull().sum()/train.shape[0]*100)
print("="*100)
print(test.isnull().sum()/test.shape[0]*100)
```

```python
test["Age"]=test["Age"].dropna()
```

```python
train.duplicated().sum()
test.duplicated().sum()
```

**Drop column not need**

```python
train.columns,test.columns
```

```python
col_train=["Name","Cabin","Embarked","PassengerId"]
train=train.drop(col_train,axis=1)
col_test=["Name","Cabin","Embarked"]
test=test.drop(col_test,axis=1)
```

```python
train
```

**Encoding**

```python
from sklearn.preprocessing import LabelEncoder
categorical_columns = train.select_dtypes(include=['object']).columns.tolist()
label_encoder = LabelEncoder()
for column in categorical_columns:
    train[column] = label_encoder.fit_transform(train[column])
```

```python
categorical_columns = test.select_dtypes(include=['object']).columns.tolist()
label_encoder = LabelEncoder()
for column in categorical_columns:
    test[column] = label_encoder.fit_transform(test[column])


# Convert the columns to integer type
train[categorical_columns] = train[categorical_columns].astype(int)
test[categorical_columns] = test[categorical_columns].astype(int)
```

# **Modeling**

```python
train_data = train.drop('Survived', axis=1)
target = train['Survived']

train_data.shape, target.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC


from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
# from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import BaggingClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

**Cross Validation**

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
clf = KNeighborsClassifier(n_neighbors = 13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
# kNN Score
round(np.mean(score)*100, 2)
```

```python
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
# decisiontree Score
round(np.mean(score)*100, 2)
```

```python
clf = RandomForestClassifier(n_estimators=13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
# Random Forest Score
round(np.mean(score)*100, 2)
```

```python
clf = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
# Naive Bayes Score
round(np.mean(score)*100, 2)
```

```python
clf = SVC()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
#SVM score
round(np.mean(score)*100,2)
```

```python
clf = BaggingClassifier(n_estimators=10, random_state=0)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
#Bagging score
round(np.mean(score)*100,2)

```

```python
from sklearn.ensemble import HistGradientBoostingClassifier


clf = HistGradientBoostingClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
#HGB score
round(np.mean(score)*100,2)

```
